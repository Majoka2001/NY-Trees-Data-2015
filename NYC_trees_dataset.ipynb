```python
status_counts=df['status'].value_counts()
```

```python
import matplotlib.pyplot as plt
status_counts.plot(kind='bar')
plt.title('Distribution of Tree Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```
